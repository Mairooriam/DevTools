```python
from pathlib import Path
from typing import Optional
import time
from dataclasses import dataclass, field

@dataclass
class Ctx:
    rootPath: Path = field(default=Path('.').resolve())
    outputFileExtension: str = field(default=".xlsx")
    outputPathStr: str = field(default="output")
    dataPathStr: str = field(default='data')
    outputFile: str = field(default="file")
    customSuffix: Optional[str] = field(default=None) 

    @property
    def outputPath(self) -> Path:
        return self.rootPath / self.outputPathStr
    
    @property
    def dataPath(self) -> Path:
        return self.rootPath / self.dataPathStr

    @property
    def outputFilePath(self) -> Path:
        base_name = self.outputFile + self.outputFileExtension
        p = self.rootPath / self.outputPath / base_name
        if not p.exists():
            return p
        else:
            if self.customSuffix:
                formatted = self.customSuffix
            else:
                localTime = time.localtime()
                formatted = time.strftime('%Y_%m_%d_%X', localTime)
                formatted = formatted.replace(":", "")
            stem = p.stem  
            suffix = p.suffix 
            new_filename = f"{stem}_{formatted}{suffix}"
            return p.with_name(new_filename)
    

ctx = Ctx()


if not ctx.outputPath.exists() or not ctx.dataPath.exists():
    print(f"coulnd't find: {ctx.dataPath.resolve()}")
    print(f"coulnd't find: {ctx.outputPath.resolve()}")
    print(f"exiting..")
    #TODO: learn erorr handling in python
else:
    print(f"Place .pdf files in {ctx.dataPath.resolve()}")
    print(f"Processed .pdf files .xlsl will be places in {ctx.outputPath.resolve()}")

# file to save




# if str1 == str2:
#     #TODO: add stuff
#     localTime = time.localtime()
#     formatted = time.strftime('%Y_%m_%d_%X', localTime)
#     formatted = formatted.replace(".", "")
#     print(f"{str1}_{formatted}")




## 2026_01_02_1039


```

```python
import camelot.io as camelot

flavors = ['lattice', 'stream', 'network', 'hybrid']


ctx = Ctx()

localTime = time.localtime()
formatted = time.strftime('%Y_%m_%d_%X', localTime)
formatted = formatted.replace(".", "")
ctx.customSuffix = formatted
for flavor in flavors:
    tables = camelot.read_pdf(pdfPath, pages='1', parallel=True, flavor=flavor)
    ctx.outputFile = flavor
    tables.export(str(ctx.outputFilePath), f='excel')
    print(f"Exported {flavor} tables to {ctx.outputFilePath}")
```

```python
import camelot.io as camelot
from PIL import Image
flavors = ['lattice', 'stream', 'network', 'hybrid']


ctx = Ctx()

image = Image.open("data/tesatr.png")
image.save("data/tesatr.pdf")
localTime = time.localtime()
formatted = time.strftime('%Y_%m_%d_%X', localTime)
formatted = formatted.replace(":", "")
ctx.customSuffix = formatted
for flavor in flavors:
    tables = camelot.read_pdf(pdfPath, pages='1', parallel=True, flavor=flavor)
    if len(tables) == 0:
        print(f"No tables found for flavor: {flavor}, skipping export.")
        continue
    ctx.outputFile = flavor
    tables.export(str(ctx.outputFilePath), f='excel')
    print(f"Exported {flavor} tables to {ctx.outputFilePath}")
```
