```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Path to your PDF file
pdf_path = "data/kotiteht5.pdf"

# Path to Tesseract executable (Windows only)
# Uncomment and update the path if needed
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def extract_text_from_image_pdf(pdf_path):
    """
    Extract text from an image-based PDF using PyMuPDF, Pillow, and Tesseract OCR.
    """
    text = ""
    pdf_document = fitz.open(pdf_path)

    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        images = page.get_images(full=True)

        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            image = Image.open(io.BytesIO(image_bytes))

            # Perform OCR on the image
            page_text = pytesseract.image_to_string(image, lang="eng")
            text += f"Page {page_num + 1}, Image {img_index + 1}:\n{page_text}\n"

    pdf_document.close()
    return text

# Example usage
if __name__ == "__main__":
    extracted_text = extract_text_from_image_pdf(pdf_path)
    print("Extracted Text:")
    print(extracted_text)
    output_md_path = "output/extracted_text.md"
    with open(output_md_path, "w", encoding="utf-8") as md_file:
        md_file.write("# Extracted Text\n\n")
        md_file.write(extracted_text)
    print(f"Extracted text saved to {output_md_path}")
```
